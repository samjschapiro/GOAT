## Main Rho Ablation Plot

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

sns.set(font_scale=1.1)
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
# plt.rcParams['text.usetex'] = True
# plt.rcParams['text.latex.preamble'] = r'\usepackage{dsfont}'

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')

import warnings
warnings.filterwarnings('ignore')
colors = ['lightsteelblue', 'cornflowerblue', 'blue', 'midnightblue', 'green', 'red']
label_size = 23
legend_size = 18
x_log_scale = False
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,6), constrained_layout=True)
# Covertype
covtype_adam_005 = pd.read_csv('logs/covtype_adam_0.05_100ep.txt')
covtype_sam_001 = pd.read_csv('logs/covtype_sam_0.01_100ep.txt')
covtype_sam_002 = pd.read_csv('logs/covtype_sam_0.02_100ep.txt')
covtype_sam_005 = pd.read_csv('logs/covtype_sam_0.05_100ep.txt')
covtype_sam_010 = pd.read_csv('logs/covtype_sam_0.1_100ep.txt')
covtype_sam_020 = pd.read_csv('logs/covtype_sam_0.2_100ep.txt')

means = np.array([covtype_sam_001[' st_acc_all'].mean(), covtype_sam_002[' st_acc_all'].mean(), covtype_sam_005[' st_acc_all'].mean(), covtype_sam_010[' st_acc_all'].mean(), covtype_sam_020[' st_acc_all'].mean()])
means = 100-means
std_devs = np.array([covtype_sam_001[' st_acc_all'].std(), covtype_sam_002[' st_acc_all'].std(), covtype_sam_005[' st_acc_all'].std(), covtype_sam_010[' st_acc_all'].std(), covtype_sam_020[' st_acc_all'].std()])
rhos = [0.01, 0.02, 0.05, 0.1, 0.2]

axs[1][1].set_title('Cover Type', fontsize=12)
axs[1][1].plot(rhos, means, color=colors[1], label='SAM');
axs[1][1].fill_between(rhos, means-std_devs, means+std_devs, alpha=0.2, color=colors[1]);
axs[1][1].plot(rhos, 100-np.repeat(covtype_adam_005[' st_acc_all'].mean(), 5), linestyle='--', color='red', label='Adam')
axs[1][1].set_ylabel('Target Domain Error', fontsize=12);
axs[1][1].set_xlabel('Rho', fontsize=12)
axs[1][1].set_xticks(rhos);
axs[1][1].set_xscale('log')
axs[1][1].legend();

# Portraits
portraits_adam_005 = pd.read_csv('logs/portraits_adam_0.05_100ep.txt')
portraits_sam_001 = pd.read_csv('logs/portraits_sam_0.01_100ep.txt')
portraits_sam_002 = pd.read_csv('logs/portraits_sam_0.02_100ep.txt')
portraits_sam_005 = pd.read_csv('logs/portraits_sam_0.05_100ep.txt')
portraits_sam_010 = pd.read_csv('logs/portraits_sam_0.1_100ep.txt')
portraits_sam_020 = pd.read_csv('logs/portraits_sam_0.2_100ep.txt')

means = np.array([portraits_sam_001[' st_acc_all'].mean(), portraits_sam_002[' st_acc_all'].mean(), portraits_sam_005[' st_acc_all'].mean(), portraits_sam_010[' st_acc_all'].mean(), portraits_sam_020[' st_acc_all'].mean()])
means = 100-means
std_devs = np.array([portraits_sam_001[' st_acc_all'].std(), portraits_sam_002[' st_acc_all'].std(), portraits_sam_005[' st_acc_all'].std(), portraits_sam_010[' st_acc_all'].std(), portraits_sam_020[' st_acc_all'].std()])

axs[1][0].set_title('Portraits', fontsize=12)
axs[1][0].plot(rhos, means, color=colors[1], label='SAM');
axs[1][0].fill_between(rhos, means-std_devs, means+std_devs, alpha=0.2, color=colors[1]);
axs[1][0].plot(rhos, 100-np.repeat(portraits_adam_005[' st_acc_all'].mean(), 5), linestyle='--', color='red', label='Adam')
axs[1][0].set_ylabel('Target Domain Error', fontsize=12);
axs[1][0].set_xlabel('Rho', fontsize=12)
axs[1][0].set_xticks(rhos);
axs[1][0].set_xscale('log')
axs[1][0].legend();

# Color MNIST
color_adam_005 = pd.read_csv('logs/color_adam_0.05_100ep.txt')
color_sam_001 = pd.read_csv('logs/color_sam_0.01_100ep.txt')
color_sam_002 = pd.read_csv('logs/color_sam_0.02_100ep.txt')
color_sam_005 = pd.read_csv('logs/color_sam_0.05_100ep.txt')
color_sam_010 = pd.read_csv('logs/color_sam_0.1_100ep.txt')
color_sam_020 = pd.read_csv('logs/color_sam_0.2_100ep.txt')

means = np.array([color_sam_001[' st_acc_all'].mean(), color_sam_002[' st_acc_all'].mean(), color_sam_005[' st_acc_all'].mean(), color_sam_010[' st_acc_all'].mean(), color_sam_020[' st_acc_all'].mean()])
means = 100-means
std_devs = np.array([color_sam_001[' st_acc_all'].std(), color_sam_002[' st_acc_all'].std(), color_sam_005[' st_acc_all'].std(), color_sam_010[' st_acc_all'].std(), color_sam_020[' st_acc_all'].std()])

axs[0][0].set_title('Color Shifted MNIST', fontsize=12)
axs[0][0].plot(rhos, means, color=colors[1], label='SAM');
axs[0][0].fill_between(rhos, means-std_devs, means+std_devs, alpha=0.2, color=colors[1]);
axs[0][0].plot(rhos, 100-np.repeat(color_adam_005[' st_acc_all'].mean(), 5), linestyle='--', color='red', label='Adam')
axs[0][0].set_ylabel('Target Domain Error', fontsize=12);
axs[0][0].set_xticks(rhos);
axs[0][0].set_xscale('log')
axs[0][0].legend(loc=1);

# Rotated MNIST
rotate_adam_005 = pd.read_csv('logs/mnist_adam_0.05_100ep.txt')
rotate_sam_001 = pd.read_csv('logs/mnist_sam_0.01_100ep.txt')
rotate_sam_002 = pd.read_csv('logs/mnist_sam_0.02_100ep.txt')
rotate_sam_005 = pd.read_csv('logs/mnist_sam_0.05_100ep.txt')
rotate_sam_010 = pd.read_csv('logs/mnist_sam_0.1_100ep.txt')
rotate_sam_020 = pd.read_csv('logs/mnist_sam_0.2_100ep.txt') 

means = np.array([rotate_sam_001[' st_acc_all'].mean(), rotate_sam_002[' st_acc_all'].mean(), rotate_sam_005[' st_acc_all'].mean(), rotate_sam_010[' st_acc_all'].mean(), rotate_sam_020[' st_acc_all'].mean()])
means = 100 - means
std_devs = np.array([rotate_sam_001[' st_acc_all'].std(), rotate_sam_002[' st_acc_all'].std(), rotate_sam_005[' st_acc_all'].std(), rotate_sam_010[' st_acc_all'].std(), rotate_sam_020[' st_acc_all'].std()])

axs[0][1].set_title('Rotated MNIST (45 degrees)', fontsize=12)
axs[0][1].plot(rhos, means, color=colors[1], label='SAM');
axs[0][1].fill_between(rhos, means-std_devs, means+std_devs, alpha=0.2, color=colors[1]);
axs[0][1].plot(rhos, 100-np.repeat(rotate_adam_005[' st_acc_all'].mean(), 5), linestyle='--', color='red', label='Adam')
axs[0][1].set_ylabel('Target Domain Error', fontsize=12);
axs[0][1].set_xticks(rhos);
axs[0][1].set_xscale('log')
axs[0][1].legend();
```
