```python
import pandas as pd
sgd_colored_mnist = pd.read_csv('logs/color.txt')
sam_colored_mnist = pd.read_csv('logs/color_sam_fixed.txt')
```

```python
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(24, 8))

print('============== ADAM ==============')
import matplotlib.pyplot as plt
sgd_colored_mnist.plot.box(ax=axs[0], showmeans=True, grid=False);
axs[0].set_ylabel('Accuracy (%)', fontsize=15);
axs[0].set_title('Adam - Colored MNIST Results', fontsize=25);
print('-------------- Mean -------------')
print(sgd_colored_mnist.mean(axis=0));
print('------------ Median -------------')
print(sgd_colored_mnist.median(axis=0));


sam_colored_mnist.plot.box(ax=axs[1], showmeans=True, grid=False);
axs[1].set_ylabel('Accuracy (%)', fontsize=15);
axs[1].set_title('SAM - Colored MNIST Results', fontsize=25);
print('============== SAM ==============')
print('-------------- Mean -------------')
print(sam_colored_mnist.mean(axis=0));
print('------------ Median -------------')
print(sam_colored_mnist.median(axis=0));
```
