```python
covtype_adam_005 = pd.read_csv('logs/covtype_adam_0.05.txt')
covtype_sam_005 = pd.read_csv('logs/covtype_sam_0.05.txt')
covtype_sam_01 = pd.read_csv('logs/covtype_sam_0.1.txt')
```

```python
covtype_adam_005.mean(axis=0), covtype_adam_005.std(axis=0)
```

```python
covtype_sam_005.mean(axis=0), covtype_sam_005.std(axis=0)
```

```python
covtype_sam_01.mean(axis=0), covtype_sam_01.std(axis=0)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12,5))
axs[0].set_title('Sharpness vs. Accuracy')
axs[0].scatter(covtype_sam_005[' st_acc_all'], covtype_sam_005[' st_sharp_all']);
axs[0].scatter(covtype_adam_005[' st_acc_all'], covtype_adam_005[' st_sharp_all']);
axs[1].set_title('Rep Norm vs. Accuracy')
axs[1].scatter(covtype_sam_005[' st_acc_all'], covtype_sam_005[' rep_norm_all']);
axs[1].scatter(covtype_adam_005[' st_acc_all'], covtype_adam_005[' rep_norm_all']);
axs[2].set_title('Rep Shift vs. Accuracy')
axs[2].scatter(covtype_sam_005[' st_acc_all'], covtype_sam_005[' st_rep_shift_all']);
axs[2].scatter(covtype_adam_005[' st_acc_all'], covtype_adam_005[' st_rep_shift_all']);
print('Sharpness/Accuracy Correlation')
print('SAM', round(covtype_sam_005.corr()[' st_acc_all'].loc[' st_sharp_all'], 2))
print('Adam:', round(covtype_adam_005.corr()[' st_acc_all'].loc[' st_sharp_all'], 2))
print('Rep Norm/Accuracy Correlation')
print('SAM', round(covtype_sam_005.corr()[' st_acc_all'].loc[' rep_norm_all'], 2))
print('Adam:', round(covtype_adam_005.corr()[' st_acc_all'].loc[' rep_norm_all'], 2))
print('Rep Shift/Accuracy Correlation')
print('SAM', round(covtype_sam_005.corr()[' st_acc_all'].loc[' st_rep_shift_all'], 2))
print('Adam:', round(covtype_adam_005.corr()[' st_acc_all'].loc[' st_rep_shift_all'], 2), '\n')

print('Accuracy Comparison')
print('SAM:', round(np.mean(covtype_sam_005[' st_acc_all']), 2), 'vs', 'Adam:',
      round(np.mean(covtype_adam_005[' st_acc_all']), 2))
```
